## Include path to simulation data

```python
import sys, os
simulations_folder = os.path.join(os.path.abspath('..'), 'simulations')
```

## Create designated save folder

```python
save_folder = r'defect-position-analysis-diff-domains-figures'
if not os.path.isdir(save_folder):
    os.makedirs(save_folder)
```

## Read in Zumer figure

```python
zumer_figures_folder = os.path.join(simulations_folder, 'zumer-figures')
zumer_plot_filename = os.path.join(zumer_figures_folder, 'disclination-position-vs-time.PNG')

zumer_plot = plt.imread(zumer_plot_filename)
plt.imshow(zumer_plot)
```

```python
zumer_plot_trim_filename = os.path.join(zumer_figures_folder, 'disclination-position-vs-time-trimmed.PNG')

zumer_plot_trim = plt.imread(zumer_plot_trim_filename)
plt.imshow(zumer_plot_trim, extent=[-245.6, 2504.4, -37.8, 38.1], aspect=25)
```

## Import flowless and flow data

```python
flowless_data_filename = os.path.join(simulations_folder, 
                                      r'two-disclination-compiled-flowless-simulation-diff-domain-size-figures\\'
                                       + r'all-peak-trajectories.pickle')
with open(flowless_data_filename, 'rb') as f:
    flowless_data = dill.load(f)
```

```python
t_1x = flowless_data['t']
x_1x = flowless_data['peak_pos_x']
y_1x = flowless_data['peak_pos_y']

t_2x = flowless_data['t_2x']
x_2x = flowless_data['peak_pos_x_2x']
y_2x = flowless_data['peak_pos_y_2x']

t_3x = flowless_data['t_3x']
x_3x = flowless_data['peak_pos_x_3x']
y_3x = flowless_data['peak_pos_y_3x']

t_zx = flowless_data['t_zx']
x_zx = flowless_data['peak_pos_x_zx']
y_zx = flowless_data['peak_pos_y_zx']
```

## Truncate data to stop after defects annihilate

```python
annihilate_1x_idx = np.where(x_1x[0, :] >= 0)[0][0]
t_1x_trunc = t_1x[:annihilate_1x_idx]
x_1x_trunc = x_1x[:, :annihilate_1x_idx]
y_1x_trunc = y_1x[:, :annihilate_1x_idx]

annihilate_2x_idx = np.where(x_2x[0, :] >= 0)[0][0]
t_2x_trunc = t_2x[:annihilate_2x_idx]
x_2x_trunc = x_2x[:, :annihilate_2x_idx]
y_2x_trunc = y_2x[:, :annihilate_2x_idx]

annihilate_3x_idx = np.argmax(x_3x[0, :])
t_3x_trunc = t_3x[:annihilate_3x_idx]
x_3x_trunc = x_3x[:, :annihilate_3x_idx]
y_3x_trunc = y_3x[:, :annihilate_3x_idx]

annihilate_zx_idx = np.argmax(x_zx[0, :])
t_zx_trunc = t_zx[:annihilate_zx_idx]
x_zx_trunc = x_zx[:, :annihilate_zx_idx]
y_zx_trunc = y_zx[:, :annihilate_zx_idx]
```

## Plot simulation data over Zumer data

```python
fig, ax = plt.subplots()
ax.imshow(zumer_plot_trim, extent=[-245.6, 2504.4, -37.8, 38.1], aspect=25)

ax.plot(t_1x_trunc, -x_1x_trunc[0, :], c='C0', lw=1, label='+1/2 140x140')
ax.plot(t_1x_trunc, -x_1x_trunc[1, :], c='C0', ls='--', lw=1, label='-1/2 140x140')
ax.plot(t_2x_trunc, -x_2x_trunc[0, :], c='C1', lw=1, label='+1/2 280x280')
ax.plot(t_2x_trunc, -x_2x_trunc[1, :], c='C1', ls='--', lw=1, label='-1/2 280x280')

ax.legend()
ax.set_title("Defect positions vs. time diff domain sizes")
ax.set_xlabel(r'$t/\tau$')
ax.set_ylabel(r'$x/\xi$')

plot_filename = os.path.join(save_folder, 'defect-pos-vs-time-280-280.png')
fig.savefig(plot_filename)
```

```python
fig, ax = plt.subplots()
ax.imshow(zumer_plot_trim, extent=[-245.6, 2504.4, -37.8, 38.1], aspect=25)

ax.plot(t_1x_trunc, -x_1x_trunc[0, :], c='C0', lw=1, label='+1/2 140x140')
ax.plot(t_1x_trunc, -x_1x_trunc[1, :], c='C0', ls='--', lw=1, label='-1/2 140x140')
ax.plot(t_zx_trunc, -x_zx_trunc[0, :], c='C2', lw=1, label='+1/2 233x233')
ax.plot(t_zx_trunc, -x_zx_trunc[1, :], c='C2', ls='--', lw=1, label='-1/2 233x233')


ax.legend()
ax.set_title("Defect positions vs. time diff domain sizes")
ax.set_xlabel(r'$t/\tau$')
ax.set_ylabel(r'$x/\xi$')

plot_filename = os.path.join(save_folder, 'defect-pos-vs-time-233-233.png')
fig.savefig(plot_filename)
```

```python
fig, ax = plt.subplots()
ax.imshow(zumer_plot_trim, extent=[-245.6, 2504.4, -37.8, 38.1], aspect=25)

ax.plot(t_1x_trunc, -x_1x_trunc[0, :], c='C0', lw=1, label='+1/2 140x140')
ax.plot(t_1x_trunc, -x_1x_trunc[1, :], c='C0', ls='--', lw=1, label='-1/2 140x140')
ax.plot(t_3x_trunc, -x_3x_trunc[0, :], c='C2', lw=1, label='+1/2 420x420')
ax.plot(t_3x_trunc, -x_3x_trunc[1, :], c='C2', ls='--', lw=1, label='-1/2 420x420')


ax.legend()
ax.set_title("Defect positions vs. time diff domain sizes")
ax.set_xlabel(r'$t/\tau$')
ax.set_ylabel(r'$x/\xi$')

plot_filename = os.path.join(save_folder, 'defect-pos-vs-time-420-420.png')
fig.savefig(plot_filename)
```
