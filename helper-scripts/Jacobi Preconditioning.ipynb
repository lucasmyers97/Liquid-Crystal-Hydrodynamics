```python
import numpy as np
```

```python
def lambda_2(N):
    lam = np.zeros(((N - 1), (N - 1)))
    for i in range(N - 1):
        for j in range(N - 1):
            if i == j:
                lam[i, j] = -2
            elif i == j + 1:
                lam[i, j] = 1
            elif i == j - 1:
                lam[i, j] = 1
                
    return lam
```

```python
def lambda_4(N):
    return np.dot(lambda_2(N), lambda_2(N))
```

```python
def A(Nx, Ny):
    return np.kron(np.eye(Nx - 1), lambda_4(Ny)) + np.kron(lambda_4(Nx), np.eye(Ny - 1)) + 2*np.kron(lambda_2(Nx), lambda_2(Ny))
```

```python
def v(N):
    M = np.zeros(((N - 1), 2), dtype=np.double)
    M[0, 0] = 1
    M[-1, -1] = 1
    return M
```

```python
def V1(Nx, Ny):
    return np.kron(v(Nx), np.eye(Ny - 1))
```

```python
def V2(Nx, Ny):
    return np.kron(np.eye(Nx - 1), v(Ny))
```

```python
def V(Nx, Ny):
    return np.hstack((V1(Nx, Ny), V2(Nx, Ny)))
```

```python
def F(N):
    M = np.zeros((N - 1, N - 1))
    for i in range(N - 1):
        for j in range(N - 1):
            M[i, j] = np.sqrt(2/N)*np.sin(np.pi*(i + 1)*(j + 1)/N)
            
    return M
```
